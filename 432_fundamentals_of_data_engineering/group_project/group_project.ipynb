```python
import json
from shapely.geometry  import shape, Point
```

```python
# test using TNP 2018-2022 data, trip 03a2ac30a46af881e6a2e6af06a3a779b67c0802
# note that the order for this dataset is (long,lat)

with open('Boundaries - Neighborhoods.geojson') as f:
    js = json.load(f)

# construct point based on lon/lat returned by geocoder
point = Point( -87.6318639497,41.8920421365)

# check each polygon to see if it contains the point
for feature in js['features']:
    polygon = shape(feature['geometry'])
    if polygon.contains(point):
        print('Point is in: ', feature['properties']['pri_neigh'])
```
