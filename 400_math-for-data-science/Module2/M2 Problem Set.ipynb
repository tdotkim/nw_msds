```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
# Use sympy package to display the equation
sym.init_printing()
x,y = sym.symbols('x,y')

solns = sym.solve([
    2*x + 1*y - 1,   # Set equations equal to 0 for solving
    3*x + 3*y + 3],
    [x, y])
print(f"Intersectiion is ({solns[x]}, {solns[y]}) ")


# Eq 1
# 2x + y = 1
# y = -2x + 1
# Eq 2
# 3x + 3y = - 3
# 3y = -3x - 3
# y = -x - 1

x_vals = np.arange(0, 10)
y1 = -2 * x_vals + 1  # Eq 1
y2 = -x_vals - 1  # Eq 2

plt.plot(x_vals, y1)
plt.plot(x_vals, y2)
plt.show()
```

<div class="alert alert-block alert-success"><b>Problem 9:</b></div>

```python
rhs = [182, 126736, 0]
rhs = np.matrix(rhs)
rhs = np.transpose(rhs)
print('\nRight Hand Side of Equation')
print(rhs)

A = [[1, 1, 1],
     [334, 959, 1280],
     [1, -2, 0]]
A = np.matrix(A)
print('\nMatrix A')
print(A)
IA = np.linalg.inv(A)

result = np.dot(IA, rhs)
result = np.rint(result)
result
```

<div class="alert alert-block alert-success"><b>Problem 10:</b></div>

```python
blends = [[.65, .6, .95], 
          [.35, .4, .05]]
A = np.matrix(blends)
print('\nMatrix A')
print(A)


print('\nMatrix B')
demands = [[15, 9],
           [2, 2],
           [25, 31]]

B = np.matrix(demands)
print(B)

prod = np.dot(A, B)
print('\nProduct')
print(prod)

# Because these are in thousands of pounds
thous_prod = 1000 * prod

prices = [4.28, 0.69]
cost = np.dot(prices, thous_prod)
print("\nCost")
print(cost)
```

<div class="alert alert-block alert-success"><b>Problem 12:</b></div>

```python
lhs = np.matrix([[5, -30], [-1, 7]])
rhs = [1, 3]

print("Coefficient Matrix")
print(lhs)

print("Inverse of it is")
inv_lhs = np.linalg.inv(lhs)
print(inv_lhs)

sol = np.dot(inv_lhs, rhs)
print(f"The solutions are {sol}")
```
